# 1) Understanding Problem Statement
---

**The KaggleX Skill Assessment Challenge involves using Used Car Price Prediction Dataset.**

#### Problem Statement:

The used car market in India is a dynamic and ever-changing landscape. Prices can fluctuate wildly based on a variety of factors including the make and model of the car, its mileage, its condition and the current market conditions. As a result, it can be difficult for sellers to accurately price their cars.

#### Approach:

We propose to develop a machine learning model that can predict the price of a used car based on its features. The model will be trained on a dataset of used cars that have been sold. The model will then be able to be used to predict the price of any used car, given its features.

#### Objective:

To build suitable Machine Learning Model for Used Car Price Prediction.

#### Benefits:

The benefits of this solution include:

- Sellers will be able to more accurately price their cars which will help them to sell their cars faster and for a higher price.
- Buyers will be able to find cars that are priced more competitively.
- The overall used car market will become more efficient.

**We believe that this project has the potential to make a significant impact on the used car market. We are excited to work on this project and to see the positive impact that it can have.**

#### Data Description

- **brand:** Brand Name of the particular car.
- **model:** Exact model name of the car of a particular brand.
- **model_year:** Manufacturing year of the vehicle crucial for assessing depreciation and technology advancements.
- **mileage:** It is usually the number of kilometer the car runs per litre. However, seesing the values it can be inferred as total distance travlled by the vehicle in miles. Key indicator of wear and tear potential maintenance requirements.
- **fuel_type:** Fuel used in the used car whether it's gasoline, diesel, electric or hybrid.
- **engine:** Engine specifications shedding light on performance and efficiency
- **transmission:** Transmission used in the used car whether automatic, manual or another variant.
- **ext_col:** Aesthetic aspectic- Exterior color
- **int_col:** Aesthetic aspectic- Interior color
- **accident:** Vehicle history of accidents or damage crucial for informed decision making.
- **clean_title:** It means that vehicle has never been deemed a total loss or experienced any severe damage due to accident, flood, fire etc.
- **price:** The listed sale price of the vehicle to help in price comparison and budgeting
